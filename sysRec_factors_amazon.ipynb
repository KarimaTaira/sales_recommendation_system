## Training the model

```python
# Pivot to summarise and count
mtrx_df = data_subset.pivot(index = 'user_id', columns ='product_id', values = 'rating').fillna(0)
mtrx_df.head()
```

```python
import numpy as np
# Demean the data
mtrx = mtrx_df.to_numpy()
ratings_mean = np.mean(mtrx, axis = 1)
normalized_mtrx = mtrx - ratings_mean.reshape(-1, 1)

normalized_mtrx
```

```python
from scipy.sparse.linalg import svds
k = 50  # You can choose a different value based on experimentation
U, sigma, Vt = svds(normalized_mtrx, k=k)
```

```python
sigma = np.diag(sigma)
```

```python
all_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + ratings_mean.reshape(-1, 1)
```

```python
preds_df = pd.DataFrame(all_predicted_ratings, columns=mtrx_df.columns, index=mtrx_df.index)
```

```python
preds_df.head()
```

```python
def recommend_products(user_id, preds_df, data, num_recommendations=5):
    # Get user's predicted ratings
    user_row = preds_df.loc[user_id]
    user_ratings = pd.DataFrame({'product_id': preds_df.columns, 'predicted_rating': user_row.values})

    # Get products the user has not rated
    user_data = data[data['user_id'] == user_id]
    products_not_rated = data[~data['product_id'].isin(user_data['product_id'])]['product_id'].unique()

    # Merge with predicted ratings
    recommendations = user_ratings[user_ratings['product_id'].isin(products_not_rated)]

    # Sort by predicted rating
    recommendations = recommendations.sort_values(by='predicted_rating', ascending=False)

    # Get top N recommendations
    top_recommendations = recommendations.head(num_recommendations)

    return top_recommendations
```

```python
user_id_to_recommend = 'AE22MK2NXQD3ZARLIOL3SLD4GU6A'
recommended_products = recommend_products(user_id_to_recommend, preds_df, data, num_recommendations=5)

print(f"Top 5 recommended products for user {user_id_to_recommend}:\n")
print(recommended_products)
```
